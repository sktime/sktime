```python
# Model Comparison in Time Series Forecasting (sktime)

This notebook demonstrates how to compare multiple forecasting models using MAE.
```

```python
from sktime.datasets import load_airline
from sktime.forecasting.theta import ThetaForecaster
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
from sktime.forecasting.naive import NaiveForecaster
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.base import ForecastingHorizon
from sktime.performance_metrics.forecasting import mean_absolute_error
```

```python
y = load_airline()

y_train, y_test = temporal_train_test_split(y, test_size=6)
```

```python
models = {
    "Theta": ThetaForecaster(sp=12),
    "Naive": NaiveForecaster(strategy="last"),
    "ExpSmoothing": ExponentialSmoothing(),
}
```

```python
results = {}

fh = ForecastingHorizon(y_test.index, is_relative=False)

for name, model in models.items():
    model.fit(y_train)
    preds = model.predict(fh)

    mae = mean_absolute_error(y_test, preds)
    results[name] = mae

results
```

```python
best_model = min(results, key=results.get)

print("Best model:", best_model)
print("MAE:", results[best_model])
```
