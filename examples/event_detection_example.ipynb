# Time Series Event Detection Example

This notebook demonstrates a simple event detection pipeline using time series data.

We simulate sensor data and use basic feature extraction (RMS) to identify abnormal behavior.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# simulate time series data
np.random.seed(42)

normal_signal = np.random.normal(0, 1, 200)
event_signal = np.random.normal(0, 5, 50)  # higher amplitude = event

signal = np.concatenate([normal_signal, event_signal, normal_signal])

df = pd.DataFrame({"signal": signal})

df.head()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df["signal"])
plt.title("Simulated Sensor Signal with Event")
plt.show()
```

```python
# simple RMS calculation
window_size = 20
rms_values = []

for i in range(len(signal) - window_size):
    window = signal[i:i+window_size]
    rms = np.sqrt(np.mean(window**2))
    rms_values.append(rms)

rms_df = pd.DataFrame({"rms": rms_values})
```

```python
plt.figure(figsize=(10, 4))
plt.plot(rms_df["rms"])
plt.title("RMS Values (Event Detection Feature)")
plt.show()
```

## Observation

The RMS values increase significantly during the event region due to higher signal amplitude.

This demonstrates how simple feature extraction can help in detecting anomalies or events in sensor data.
