```python
# Forecasting Example using sktime
```

```python
from sktime.datasets import load_airline
from sktime.forecasting.naive import NaiveForecaster
```

```python
# Load dataset
y = load_airline()
y.head()
```

```python
# Initialize model
forecaster = NaiveForecaster(strategy="last")
```

```python
# Train model
forecaster.fit(y)
```

```python
# Forecast next 3 steps
y_pred = forecaster.predict(fh=[1, 2, 3])
y_pred
```

```python
# Explanation:
# NaiveForecaster predicts future values based on last observed value
# fh = forecasting horizon (future steps)
```
