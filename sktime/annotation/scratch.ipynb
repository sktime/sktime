```python
member = np.array(range(X.shape[0]))
# member[1] = 0
alpha = 1
```

```python
u = np.sort(np.unique(member))  # unique array of cluster labels
N = len(u)  # number of clusters

for i in range(N):  # relabel clusters to be consecutive numbers (when user specified)
    member[np.where(member == u[i])[0]] = i

# sizes = np.zeros(2*N) #FIXME: append to this list
sizes = [sum(member == i) for i in range(N)]  # calculate initial cluster sizes

# array of within distances
```

```python
def get_within(X, alpha):
    n = X.shape[0]
    return list(
        sum(
            np.power(
                np.sqrt(
                    sum(
                        (
                            X.iloc[
                                i,
                            ]
                            - X.iloc[
                                j,
                            ]
                        )
                        * (
                            X.iloc[
                                i,
                            ]
                            - X.iloc[
                                j,
                            ]
                        )
                    )
                ),
                alpha,
            )
            for j in range(n)
        )
        / (n**2)
        for i in range(n)
    )


within = [
    get_within(
        X.loc[
            [
                i,
            ]
        ],
        alpha,
    )[0]
    for i in range(N)
]
```

```python
def get_between(X, Y, alpha):
    n = X.shape[0]
    m = X.shape[0]
    return list(
        sum(
            np.power(
                np.sqrt(
                    sum(
                        (
                            X.iloc[
                                i,
                            ]
                            - Y.iloc[
                                j,
                            ]
                        )
                        * (
                            X.iloc[
                                i,
                            ]
                            - Y.iloc[
                                j,
                            ]
                        )
                    )
                ),
                alpha,
            )
            for j in range(m)
        )
        / (m * n)
        for i in range(n)
    )
```

```python
D = pd.DataFrame(index=range(N), columns=range(N))

for i in range(N):
    for j in range(N):
        between = get_between(
            X.loc[
                [
                    i,
                ]
            ],
            X.loc[
                [
                    j,
                ]
            ],
            alpha,
        )[0]
        D.iloc[i, j] = D.iloc[j, i] = 2 * between - within[i] - within[j]
```
